## Overview

We examine the semantic structure of BlueSky's left-leaning feed ecosystem using three computational diagnostics:

1. Inter-feed centroid similarity
2. Within-feed dispersion (semantic cohesion)
3. Principal Component Analysis (PCA) of feed centroids  

These methods allow us to evaluate: **ideological cohesion, sub-community clustering, and the dimensional structure of discourse.**


## Inter-Feed Semantic Similarity

**Mean inter-feed similarity:** 0.487  
**Standard deviation:** 0.227  
**Range:** −0.133 to 0.993  


### Interpretation

A mean similarity of ~0.49 indicates **moderate semantic proximity but not homogeneity**. For reference:

- > 0.80 → Strong structural cohesion  
- 0.60–0.80 → Moderate clustering  
- < 0.60 → Distributed ecosystem  

The observed value (0.487) places the left ecosystem in the **distributed regime**, suggesting meaningful variation across feeds.

The relatively high standard deviation (0.227) indicates significant dispersion across pairs:

- Some feeds are nearly identical (0.99 similarity), indicating micro-echo clusters.
- Some feed pairs are negatively aligned (−0.13), reflecting substantial semantic divergence.

**Conclusion:** The left feed ecosystem is not monolithic. It contains tightly aligned sub-clusters embedded within a broader heterogeneous structure.

---

## II. Within-Feed Dispersion (Semantic Cohesion)

**Mean dispersion:** 0.489  
**Range:** 0.000 to 0.741  

Dispersion is defined as:

\[
\text{Dispersion} = 1 - \text{mean cosine similarity to feed centroid}
\]

- Low dispersion → tightly framed, echo-like feed  
- High dispersion → broad thematic diversity  

### Interpretation

An average dispersion near 0.49 suggests **moderate internal diversity** within feeds.

However, the wide range is informative:

- Some feeds exhibit very low dispersion (~0.00), indicating highly repetitive or tightly focused discourse.
- Others approach 0.74, indicating substantial topic diversity.

This demonstrates that echo chamber dynamics are **feed-specific**, not ecosystem-wide.

**Conclusion:** The ecosystem includes both tightly cohesive activist nodes and broader aggregation hubs.

---

## III. PCA: Dimensional Structure of the Ecosystem

**PC1 explained variance:** 12.2%  
**PC2 explained variance:** 8.2%  
**Total (PC1 + PC2):** 20.4%

PCA identifies dominant axes of semantic variation across feed centroids.

### Interpretation

Only ~20% of total variance is captured in the first two principal components. This indicates:

- No single dominant ideological axis
- Multi-dimensional semantic organization
- Variation distributed across multiple latent thematic dimensions

In contrast, strongly polarized ecosystems often exhibit a dominant PC1 (>40% variance).

**Conclusion:** The left ecosystem is structurally multi-dimensional rather than organized along a single ideological gradient.

---

## IV. Structural Interpretation

Taken together:

- Moderate inter-feed similarity (0.487)
- Substantial pairwise variance
- Wide dispersion range
- Multi-dimensional PCA structure

These findings suggest that the left-leaning feed ecosystem is **structurally heterogeneous**, containing:

- Highly aligned micro-clusters
- Thematically differentiated sub-communities
- No dominant organizing semantic axis

This challenges simplified echo chamber narratives and instead supports a model of **networked pluralism within ideological communities**.

---

## V. Methodological Caveats

- Analysis is restricted to a two-week window.
- Only left-leaning feeds are included.
- No cross-ideological comparison group is present.

Therefore, conclusions are limited to **intra-ideological structure**, not polarization.

---

## Summary

While tightly aligned sub-clusters exist, the broader left-leaning feed ecosystem demonstrates meaningful internal diversity and multi-dimensional semantic organization. The data do not support a model of monolithic discourse, but rather a pluralistic and structurally differentiated ideological network.
