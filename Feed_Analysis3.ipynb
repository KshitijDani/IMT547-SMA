# Analysis of Semantic Structure in Left-Leaning Political Feeds

We examine the semantic structure of up to 1,000 recent posts collected from **77 left-leaning political feeds** by embedding them using a language model. We explore how similar the feeds are to one another and how diverse the conversations are within each feed.

Three main metrics were used:

- **Feed dispersion** — how internally diverse the posts in a feed are
- **Feed similarity** — how semantically similar different feeds are to each other
- **Principal Component Analysis (PCA)** — a dimensionality reduction technique used to identify broad thematic patterns across feeds

These methods allow us to evaluate: **ideological cohesion, sub-community clustering, and the dimensional structure of discourse.**

---

# Feed Dispersion: How Diverse Are Conversations Within Feeds?

Dispersion measures the **semantic spread of posts within a single feed**. Lower values indicate that posts tend to focus on a narrow topic, while higher values suggest that the feed covers a wider range of themes.

## Summary Statistics

| Metric | Value |
|---|---|
| Mean dispersion | ~0.51 |
| Median dispersion | ~0.53 |
| Approximate range | 0–0.71 |

- Nearly identical feed pairs indicate micro-echo clusters.
  - EX: Democracy vs. Trump/MAGA <-> TrumpWatch (similarity = 0.9857)

- Least similar feed pairs reflect significant semantic divergence.
  - EX: Democracy against AfD <-> Texas Dems (similarity = -0.0734)

This suggests that even though these feeds share a general ideological orientation, the content within them is **not tightly constrained to a single topic or narrative**.

Instead, many feeds likely mix several forms of political discussion, such as:

- electoral politics
- policy debates
- activism and advocacy
- commentary and opinion
- news sharing

**Conclusion:** 

The left feed ecosystem is not monolithic.  For example, Anti-Trump focused feeds form tightly aligned semantic clusters with nearly identical centroid strucuture (cosine ~0.99).  Most left leaning feeds are tightly aligned sub-clusters embedded within a broader heterogeneous structure.

# PCA Results: Mapping the Structure of Political Conversation

**Principal Component Analysis (PCA)** was used to project the high-dimensional embeddings into a two-dimensional space.

This allows us to visualize **broad thematic differences between feeds**.


**PC1 explained variance:** 11.6%  
**PC2 explained variance:** 8.0%  
**Total (PC1 + PC2):** 19.6%

### PC1 Extremes

| Feed | PC1 |
|-----|-----|
| Memes of the Resistance | -0.437 |
| Democracy against AfD | -0.265 |
| Texas Dems | 0.326 |
| House Democrats | 0.537 |

### PC2 Extremes

| Feed | PC2 |
|-----|-----|
| The Democrats | -0.239 |
| NC Progressives | -0.189 |
| Democrats on impeachment | 0.247 |
| Democratic Senators | 0.735 |

Only ~20% of the total variance is captured by the first two principal components.

This indicates that no single semantic dimension dominates the embedding space. Instead, variation across feeds is distributed across multiple thematic directions.

Feeds at the extremes of **PC1** represent different styles of political discourse. One interpretation is a spectrum between **institutional politics and grassroots activism**, or between **policy-focused discussion and personality-driven commentary**. This suggests the axis captures differences in *how political issues are discussed*.

The **second principal component (PC2)** highlights more distinct subcommunities. One feed appears as a strong outlier (~0.73), which idicares discourse that is semantically different from most others.

**Conclusion:**\
The left-leaning feed ecosystem appears structurally multi-dimensional, with multiple discourse styles and thematic subcommunities rather than a single dominant ideological gradient.

---

# Key Takeaway

The most important result is the **moderate median dispersion (~0.53)** across feeds.

This indicates that while the feeds share an ideological orientation, the conversations within them are **not highly uniform**.

Instead of a tightly bounded echo chamber, the data suggests a **political discourse ecosystem** composed of multiple overlapping topics and communities.

In short, ideological alignment does not necessarily imply **semantic uniformity**.
